```python
# Cwiczenie 1
import matplotlib.pyplot as plt
import numpy as np

# y - 2x - 1
# xpoints = [-3, -2, -1, 0, 1, 2, 3]

# # 2x -1

# ypoints = [2*x -1 for x in xpoints]

# xpoints = np.array([-3, -2, -1, 0, 1, 2, 3])
# ypoints = 2*xpoints -1
xpoints = np.arange(-3, 3, 1)
ypoints = 2*xpoints -1

plt.plot(xpoints, ypoints, c='m', marker='o', linestyle='--', label='y-2x-1')
plt.title("wykres funkcji liniowej")
plt.legend()
plt.xlabel("OŚ X")
plt.ylabel("OŚ Y")
plt.grid()
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np
xpoints = np.linspace(-5,5,100)
ypoints = 2 * np.square(xpoints) -3
plt.plot(xpoints, ypoints, c='m', marker='o', linestyle=':', label=' 2x^2 -3')
plt.title("wykres funkcji kwadratowej")
plt.legend()
plt.xlabel("OŚ X")
plt.ylabel("OŚ Y")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.random.rand(25)
y = np.random.rand(25)

plt.scatter(x, y, color='blue', marker='o', label="losowe punkty")
plt.title("Losowe punkty")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import matplotlib.pyplot as plt

przedmioty = ["matma", "wf", "j.polski", "geografia", "j.angielski", "infa"]

procenty = [20, 15, 25, 10 , 10, 20]

plt.pie(procenty, labels=przedmioty, autopct='%1.1f%%',
colors=["red", "blue", "green", "orange", "purple", "magenta"])
plt.title("Przedmioty")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap="viridis")
plt.show()
# plt.savefig("wykres3D.png")
```
