```python
s = pd.Series([18, 17, 19, 20, "kot"], index=["Kuba", "Kasia", "Wojtek", "Adam", "Tomek"])
print(s)

print(f"rozmiar: {s.size}")
print(f"kształt: {s.shape}")
print(f"typ danych: {s.dtype}")
print(f"indexy: {s.index}")
print(f"wartości: {s.values}")
print(f"wartość dla konkretnego indexu: {s["Wojtek"]}")
print(f"wartość dla elementu na pozycji 1: {s.iloc[1]}")
```

```python
import pandas as pd

df = pd.read_excel("/content/Gigathon_Wyniki.xlsx")
```

```python
df
```

```python
print(df.head(5))
print("\n")
print(df.tail(5))
```

```python
df['Wynik procentowy'] = df["Punktacja"]/120*100
df
```

```python
# Do usunięcia rekordów z wartością NaN służy metoda .dropna().
# Parametr inplace informuje że usuwamy dane z oryginalnego
# dataframe, domyślnie metoda zwraca kopię dataframe z usuniętymi
# rekordami.
df.dropna(inplace=True)
df
```

```python
# Do sortowania wartości służy metoda .sort_values().
# Parametry:
# 'by' - według której kolumny odbywa się sortowanie,
# 'ascending' - kolejność sortowania (rosnąca lub malejąca).
df.sort_values(by="Punktacja", ascending=False, inplace=True)
df
```

```python
df["Zakwalifikowany"] = df["Wynik procentowy"]>70
df
```

```python
df.groupby(["Zakwalifikowany", "Wiek"])[["Punktacja", "Wynik procentowy"]].agg(["count", "min", "max", "mean", "median"])
```
